Biblioteca para analise de dados

```python
import pandas as pd  
```

Criaçao de tabela para dados em csv

```python
notas = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.3/ratings.csv')
```

Quantidade de linhas e colunas 

```python
notas.shape
```

Modificando o nome das colunas

```python
notas.columns = ['usuarioid', 'filme', 'nota', 'momento']
notas

```

Valores existentes na tabela

```python
notas['nota'].unique()
```

Quantidade de votos por numero individual

```python
notas['nota'].value_counts()
```

Media e mediana das notas em geral

```python
print(notas['nota'].mean())
print(notas['nota'].median())
```

```python
notas.nota.head()
```

Plotagem das notas mais votadas

```python
notas.nota.plot(kind='hist');
```
